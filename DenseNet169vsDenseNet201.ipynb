```python
import numpy as np
import keras
import tensorflow as tf
from tensorflow.keras.applications.densenet import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
```

```python
# DenseNet169
```

```python
D169 = tf.keras.applications.DenseNet169(
    include_top=True,
    weights='imagenet',
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000,
    classifier_activation='softmax'
)
```

```python
img_path = './cat.jfif'
img = keras.utils.load_img(img_path, target_size=(224, 224, 3))
x = keras.utils.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
D169_preds = D169.predict(x)

D169_results = decode_predictions(D169_preds, top=3)[0]

for i, (id, label, prob) in enumerate(D169_results):
    print(f"{i+1}: {label} ({prob:.2%})")
```

```python
D201_preds = D201.predict(x)

D201_results = decode_predictions(D201_preds, top=3)[0]

for i, (id, label, prob) in enumerate(D201_results):
    print(f"{i+1}: {label} ({prob:.2%})")
```
